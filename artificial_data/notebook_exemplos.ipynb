```python
# Caso de uso: default da classe
from maker_dataset_command import DatasetGenerator
obj = DatasetGenerator()
X, y, y_train = obj.get_arrays()
df = obj.get_dataframe()
obj.plot(y='all')
obj.plot(y='train')
```

```python
# Caso de uso: alterando a similaridade dos grupos - Menores valores indicam mais dispersão, depende de n_points
from maker_dataset_command import DatasetGenerator
obj = DatasetGenerator(similarity=[1])
obj.plot(y='all')
obj = DatasetGenerator(similarity=[.2])
obj.plot(y='all')
obj = DatasetGenerator(similarity=[.02])
obj.plot(y='all')
```

```python
# Caso de uso: alterando a dissimilaridade dos grupos, quanto mais proximo de 1 mais distantes, mais proximos de 0 mais proximos
from maker_dataset_command import DatasetGenerator
obj = DatasetGenerator(dissimilarity=[.99])
obj.plot(y='all')
obj = DatasetGenerator(dissimilarity=[.09])
obj.plot(y='all')
```

```python
# Caso de uso: Diferentes n_points
from maker_dataset_command import DatasetGenerator
obj = DatasetGenerator(n_points=1000, limits=((0,0),(5000,5000)))
obj.plot(y='all')
obj = DatasetGenerator(n_points=10000, limits=((0,0),(5000,5000)))
obj.plot(y='all')
obj = DatasetGenerator(n_points=100000, limits=((0,0),(5000,5000)))
obj.plot(y='all')
```

```python
# Caso de uso: Diferentes n_groups
from maker_dataset_command import DatasetGenerator
obj = DatasetGenerator(limits=((0,0),(1000,1000)), n_groups=3, binary_groups=[0,0,1],
                       prop_groups=[1/3], similarity=[.02], dissimilarity=[.8])
obj.plot(y='all')
obj = DatasetGenerator(limits=((0,0),(1000,1000)), n_groups=5, binary_groups=[0,0,1,1,0],
                       prop_groups=[1/5], similarity=[.02], dissimilarity=[.8])
obj.plot(y='all')
obj = DatasetGenerator(limits=((0,0),(1000,1000)), n_groups=8, binary_groups=[0,0,1,1,0,0,1,1],
                       prop_groups=[1/8], similarity=[.02], dissimilarity=[.8])
obj.plot(y='all')
```
